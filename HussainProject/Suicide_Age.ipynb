```python
#Importing dependencies
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#assigning csv to a variable and then reading into it and producing the first 5 rows in dataframe.
file = "suicide_age.csv"
age_df = pd.read_csv(file)
age_df.head()
```

```python
#sorting dataframe in order to display United States only. 
is_United_States =  age_df['Entity']=='United States'
df = age_df[age_df['Entity']=='United States']
df.head()
```

```python
#Sorting values again by year
df.sort_values(by='Year', ascending=False)
```

```python
#Graph specs, assigning multiple lines on graph
ax = plt.gca() 
df.plot(kind='line', x='Year', y='70+ years old (per 100,000)', color='purple', ax=ax, linestyle='solid', linewidth=1.5, fontsize=8.5, figsize=(8,4))
df.plot(kind='line', x='Year', y='50-69 years old (per 100,000)', color='red', linewidth=1.5, ax=ax, linestyle='solid')
df.plot(kind='line', x='Year', y='15-49 years old (per 100,000)', color='blue', ax=ax, linestyle='solid', linewidth=1.5, fontsize=8.5, figsize=(8,4))
df.plot(kind='line', x='Year', y='5-14 years old (per 100,000)', color='green', linewidth=1.5, ax=ax, linestyle='solid')
#Customized gridline
ax.grid(linestyle=':', linewidth='0.5', color='gray')
#Assigning the title
plt.title("Suicide Rates in the U.S. 1990-2017 (by age)", fontname="Arial", fontsize=13)
#Assigning labels
plt.xlabel('Year', fontname="Arial", fontsize=9)
plt.ylabel('Percentage',fontname="Arial", fontsize=9)
#Displays graph
plt.show()
#Saves graph
plt.savefig('suicide_age.png')
```
